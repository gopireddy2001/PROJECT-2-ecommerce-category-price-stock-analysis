```python
import requests
import pandas as pd
```

```python
# fetching the data from API
url = "https://fakestoreapi.com/products"
response=requests.get(url)
data=response.json() #converts to python list or dict
```

```python
df=pd.json_normalize(data)
df.head()
```

```python
df_cleaned = df.loc[:, ['id', 'title', 'category', 'price', 'rating.rate', 'rating.count']].copy()
#df_cleaned.rename(columns={'rating.rate': 'rating', 'rating.count': 'rating_count'}, inplace=True)
df_cleaned.rename(columns={
    'rating.rate': 'rating',
    'rating.count': 'rating_count'
}, inplace=True)
```

```python
df_cleaned.to_csv("ecommerce_category_price_stock.csv", index=False)
from google.colab import files
files.download("ecommerce_category_price_stock.csv")

```
